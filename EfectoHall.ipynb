## Incertidumbre de la Multiplicación

Para este caso el manejo de incertidumbres fue de la siguiente manera:

La incertidumbre absoluta de $Z=a\cdot b$, con $\gamma a$ y $\gamma b$ las incertidumbres de a y b respectivamente, es $\gamma Z=a_0\gamma b+b_0\gamma a$.


## Incertidumbre de las Lecturas

La incertidumbre asociada a las lecturas son:

$A[mA]$ es $\pm0.012A_0 [mA]$

$B[kG]$ es $\pm0.005B_0 [kG]$

$V[mV]$ es $\pm0.008V_0 [mV]$



## Incertidumbre de la Pendiente

La incertidumbre de la pendiente se calculó por medio de pendientes alternadas entre los puntos es decir:

Si $(x,y)=[a,b,c,d,e,f]$ donde cada elemento es una coordenada, entonces:

$m     =  pendiente(e,a)$ 

$m_{max} =  pendiente(d,b)$

$m_{min} =  pendiente(f,c)$

Se calculó $|m_{max} - m|$ y $|m_{min} - m|$ , se compararon y se tomó como incertidumbre el valor máximo entre estos dos.
 


Por medio de la ecuación (6) se puede deducir que el Coeficiente Hall de la punta de prueba utilizada esta dada por $R_H=\frac{\omega V_H}{I_H B}$.

De la gráfica anterior se obtuvo la relación $V_H=I_HBm$. 

Utilizando estas dos ecuaciones con los valores correspondientes se obtiene que $R_H=\omega m$
