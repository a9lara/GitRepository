## Del artículo de Yoshikawa & Sasaki (2006) voy a graficar la evolución temporal de la densidad bariónica 

$n_{\mathrm{H,0}}$ es la densidad numérica de hidrógeno actual.

$n_{\mathrm{H,0}}=2.2\times10^{-7} \mathrm{cm}^{-3}$.

$T_\mathrm{i}$ es la temperatura inicial a $z_\mathrm{i}=10$.

$T_\mathrm{i}=5\times10^3 \mathrm{K}$.

La materia bariónica tiene una densidad numérica de hidrógeno  $n_\mathrm{H} =n_{\mathrm{H,0}}(1+z_\mathrm{i})^3$ y una temperatura $T_\mathrm{i}$ a un corrimiento al rojo inicial $z_\mathrm{i}=10$.



```julia
function nH(zi,nH0) # cm^(-3)
    
    zs  = Float64[]
    nHs = Float64[]    
           
    z = zi
    
    while z > 0
        
        n = nH0 * (1+z)^3
        
        push!(zs,z)
        push!(nHs,n)
        
        z = z-0.2
        
    end
    
    zs,nHs
    
end
```

```julia
zs,nHs=nH(10,2.2e-7);
```

```julia
plot(zs,log10.(nHs),"-k",linewidth=2)
plot([10,0],[log10(2.2e-7*(1+2)^3),log10(2.2e-7*(1+2)^3)],":k",label=L"n_\mathrm{H}(z=2)",linewidth=2)
annotate(L"5.6\times10^{-6} \ \mathrm{cm}^{-3}", xy=(9.9, -5.2), fontsize=20)

title("Evolución temporal de la densidad de bariones",fontsize=20)

xlabel(L"z",fontsize=20)
xticks(fontsize=15)
xlim(10, 0)


ylabel(L"\log_{10}(n_\mathrm{H} \ [\mathrm{cm}^{-3}])",fontsize=20)
yticks(fontsize=15)

legend(fontsize=20);
```
