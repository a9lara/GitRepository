```julia
using PyPlot

function TiroParabólico(q0,v0,t)
  
   ts=Float64[]
    qs=Vector{Float64}[]
    vs=Vector{Float64}[]

    while  !(q0[2]<0)
    
  # t: tamaño de paso. 
  # g: aceleración gravitatoria en la Tierra

  g=9.81 # m/s^2
  
  # Integración:
  
        v_nueva=[v0[1],v0[2]-g*t]
        q_nueva=[q0[1]+v0[1]*t,(-g*(t^2))/2+v0[2]*t+q0[2]]
        
        push!(qs,[q0])
        push!(vs,[v0])
        push!(ts,t)
        
        v0=v_nueva
        q0=q_nueva
       
    

   end
    ts,hcat(qs...)'
end


# Ahora usamos la función que acabamos de hacer poniendole condiciones iniciales y luego graficamos la trayectoria.

TS,QS=TiroParabólico([0,0],[20,80],0.001);
X1T,Y1T=QS[:,1],QS[:,2];
plot(X1T,Y1T)
xlabel("x [m]")
ylabel("y [m]")
title("Trayectoria Tiro Parabólico")
```
