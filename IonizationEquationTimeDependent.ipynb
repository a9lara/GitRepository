## $$\text{Resolver la siguiente ecuación}$$ 

## $\boxed{\ddot{n}_{\mathrm{H}_\mathrm{I}}(t)=-\dot\Gamma_{\mathrm{H}_\mathrm{I}}(t)n_{\mathrm{H}_\mathrm{I}}(t)-\Gamma_{\mathrm{H}_\mathrm{I}}(t)\dot n_{\mathrm{H}_\mathrm{I}}(t)
+\alpha_{\mathrm{rec,H}_\mathrm{II}}n_\mathrm{e}\Gamma_{\mathrm{H}_\mathrm{I}}(t)n_{\mathrm{H}_\mathrm{I}}(t)-\alpha_{\mathrm{rec,H}_\mathrm{II}}^2n_\mathrm{e}^2n_{\mathrm{H}_\mathrm{II}}(t)}$ (1)

## $n_\mathrm{e}\approx\text{cte}$

## $n_\mathrm{H}=n_{\mathrm{H}_\mathrm{I}}(t)+n_{\mathrm{H}_\mathrm{II}}(t)$

## $\text{Normalizamos haciendo}$ 

## $n_\mathrm{H}=1$

## $\text{Tal que}$

## $n_{\mathrm{H}_\mathrm{I}}(t)+n_{\mathrm{H}_\mathrm{II}}(t)=1$ 

## $\boxed{n_{\mathrm{H}_\mathrm{II}}(t)=1-n_{\mathrm{H}_\mathrm{I}}(t)}$ (2)

## $\boxed{\Gamma(t) = \left \{ \begin{matrix} \Gamma_o & \mbox{ si } & t\leq0
\\ a\Gamma_o & \mbox{ si } & t > 0\end{matrix}\right.}$ (3)

## $\boxed{\ddot{n}_{\mathrm{H}_\mathrm{I}}(t)=-\dot\Gamma_{\mathrm{H}_\mathrm{I}}(t)n_{\mathrm{H}_\mathrm{I}}(t)-\Gamma_{\mathrm{H}_\mathrm{I}}(t)\dot n_{\mathrm{H}_\mathrm{I}}(t)
+\alpha_{\mathrm{rec,H}_\mathrm{II}}n_\mathrm{e}\Gamma_{\mathrm{H}_\mathrm{I}}(t)n_{\mathrm{H}_\mathrm{I}}(t)-\alpha_{\mathrm{rec,H}_\mathrm{II}}^2n_\mathrm{e}^2+\alpha_{\mathrm{rec,H}_\mathrm{II}}^2n_\mathrm{e}^2n_{\mathrm{H}_\mathrm{I}}(t)}$ (4)

## $\text{Sea}$

## $\aleph=\alpha_{\mathrm{rec,H}_\mathrm{II}}n_\mathrm{e}$

## $\boxed{\ddot{n}_{\mathrm{H}_\mathrm{I}}(t)=-\dot\Gamma_{\mathrm{H}_\mathrm{I}}(t)n_{\mathrm{H}_\mathrm{I}}(t)-\Gamma_{\mathrm{H}_\mathrm{I}}(t)\dot n_{\mathrm{H}_\mathrm{I}}(t)
+\aleph\Gamma_{\mathrm{H}_\mathrm{I}}(t)n_{\mathrm{H}_\mathrm{I}}(t)-\aleph^2+\aleph^2n_{\mathrm{H}_\mathrm{I}}(t)}$ (5)

# $\bullet  \ t\leq0$

## $\text{Resolvemos para el caso } t\leq0 \text{, es decir, } \Gamma(t) =  \Gamma_o $ 

## $\boxed{\ddot{n}_{\mathrm{H}_\mathrm{I}}(t)=-\Gamma_o\dot n_{\mathrm{H}_\mathrm{I}}(t)
+\aleph\Gamma_on_{\mathrm{H}_\mathrm{I}}(t)-\aleph^2+\aleph^2n_{\mathrm{H}_\mathrm{I}}(t)}$ (6)

## $\text{La solución de (6) es}$

## $\boxed{{n}_{\mathrm{H}_\mathrm{I}}(t)=\frac{c_1}{e^{(\aleph+\Gamma_o)t}}+\frac{\aleph}{\aleph+\Gamma_o}+c_2e^{\aleph t}}$(7)

# $\bullet  \ t>0$

## $\text{Y resolviendo el caso para} t>0 \text{, es decir, } \Gamma(t) =  a\Gamma_o $ 

## $\boxed{{n}_{\mathrm{H}_\mathrm{I}}(t)=\frac{c_3}{e^{(\aleph+a\Gamma_o)t}}+\frac{\aleph}{\aleph+a\Gamma_o}+c_4e^{\aleph t}}$(8)

## $\therefore$

## $\boxed{{n}_{\mathrm{H}_\mathrm{I}}(t)= \left \{ \begin{matrix} \frac{c_1}{e^{(\aleph+\Gamma_o)t}}+\frac{\aleph}{\aleph+\Gamma_o}+c_2e^{\aleph t} & \mbox{ si } & t\leq0
\\ \frac{c_3}{e^{(\aleph+a\Gamma_o)t}}+\frac{\aleph}{\aleph+a\Gamma_o}+c_4e^{\aleph t} & \mbox{ si } & t > 0\end{matrix}\right.}$ (9)

## $\text{Como no queremos que (9) diverga cuando } t\rightarrow \infty \text{ o } t\rightarrow -\infty$ 

## $c_2=0$ 

## $c_3=0$

## $\boxed{{n}_{\mathrm{H}_\mathrm{I}}(t)= \left \{ \begin{matrix} \frac{c_1}{e^{(\aleph+\Gamma_o)t}}+\frac{\aleph}{\aleph+\Gamma_o} & \mbox{ si } & t\leq0
\\ \frac{\aleph}{\aleph+a\Gamma_o}+c_4e^{\aleph t} & \mbox{ si } & t > 0\end{matrix}\right.}$ (10)

## $\text{Ahora vamos a buscar } c_1 \text{ y } c_4$ 


# $$\text{Implementamos la función}$$

```julia
function nHI(tiempo,a,nHIo)
    
    # nHI se va a evaluar desde -tiempo hasta tiempo
    # a es el factor que se multiplica a la Γ para t>0
    # nHIo es la densidad de HI al tiempo t=0
    
Γ=1
ne=1
α_rec_HII=1

ts=Float64[]
nHIs=Float64[]
nHIIs=Float64[]
for t in -tiempo:0.001:tiempo
        
       if t<=0
            densidad_HI=((ne*α_rec_HII)/(ne*α_rec_HII+Γ))+((nHIo-((ne*α_rec_HII)/(ne*α_rec_HII+Γ)))*exp(t*(ne*α_rec_HII)))
            densidad_HII=1-densidad_HI
            push!(nHIIs,densidad_HII)
            push!(nHIs,densidad_HI)
            push!(ts,t)  
            else
            densidad_HI=((nHIo-((ne*α_rec_HII)/(ne*α_rec_HII+a*Γ)))/(exp(t*(ne*α_rec_HII+a*Γ))))+((ne*α_rec_HII)/(ne*α_rec_HII+a*Γ))
            densidad_HII=1-densidad_HI
            push!(nHIIs,densidad_HII)
            push!(nHIs,densidad_HI)
        push!(ts,t)   
        end
    end
    ts,nHIs,nHIIs
end
```

# $$\text{Evaluamos y graficamos}$$

```julia
using PyPlot
```

```julia
tiempo,densidad_HI,densidad_HII=nHI(10,0,1);

plot(tiempo,densidad_HI,label=L"n_{\mathrm{H}_\mathrm{I}}","-k")
plot(tiempo,densidad_HII,label=L"n_{\mathrm{H}_\mathrm{II}}",":k")

annotate(L"\Gamma(t\leq0) = 1", xy=(-9, 0.2), fontsize=20)
annotate(L"\Gamma(t>0) = 0", xy=(2, 0.2), fontsize=20)

title(L"n_{\mathrm{H}_\mathrm{I}(t=0)=0.2}",fontsize=20)

xlabel(L"t",fontsize=20)
xticks(fontsize=15)
#xlim(10, 0)


ylabel(L"n_\mathrm{H}",fontsize=20)
yticks(fontsize=15)

legend(fontsize=20);
```
