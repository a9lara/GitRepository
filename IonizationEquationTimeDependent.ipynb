## $\text{Tenemos el siguiente sistema de ecuaciones.}$ 

## $\boxed{\dot{n}_{\mathrm{H}_\mathrm{I}}(t)=-\Gamma_{\mathrm{H}_\mathrm{I}}(t)n_{\mathrm{H}_\mathrm{I}}(t)+\alpha_{\mathrm{rec,H}_\mathrm{II}}n_\mathrm{e}n_{\mathrm{H}_\mathrm{II}}(t)}$ (1)

## $\boxed{\dot{n}_{\mathrm{H}_\mathrm{II}}(t)=\Gamma_{\mathrm{H}_\mathrm{I}}(t)n_{\mathrm{H}_\mathrm{I}}(t)-\alpha_{\mathrm{rec,H}_\mathrm{II}}n_\mathrm{e}n_{\mathrm{H}_\mathrm{II}}(t)}$ (2)

## $\text{Suponemos:}$

## $n_\mathrm{e}\approx\text{cte}$

## $n_\mathrm{H}=n_{\mathrm{H}_\mathrm{I}}(t)+n_{\mathrm{H}_\mathrm{II}}(t)$

## $\text{Normalizamos haciendo}$ 

## $n_\mathrm{H}=1$

## $\text{Tal que}$

## $n_{\mathrm{H}_\mathrm{I}}(t)+n_{\mathrm{H}_\mathrm{II}}(t)=1$ 

## $\boxed{n_{\mathrm{H}_\mathrm{II}}(t)=1-n_{\mathrm{H}_\mathrm{I}}(t)}$ (3)

## $\bullet$ $\text{Caso 1}$

## $\text{Supongamo que para un tiempo menor que cero se tiene equilibrio de fotoionización.}$ 

## $\text{Tal que (1) y (2) son:}$

## $\boxed{-\Gamma_{\mathrm{H}_\mathrm{I}}(t)n_{\mathrm{H}_\mathrm{I}}(t)+\alpha_{\mathrm{rec,H}_\mathrm{II}}n_\mathrm{e}n_{\mathrm{H}_\mathrm{II}}(t)=0}$ (4)

## $\boxed{\Gamma_{\mathrm{H}_\mathrm{I}}(t)n_{\mathrm{H}_\mathrm{I}}(t)-\alpha_{\mathrm{rec,H}_\mathrm{II}}n_\mathrm{e}n_{\mathrm{H}_\mathrm{II}}(t)=0}$ (5)

## $\text{Resolviendo nos queda:}$ 

## $\boxed{n_{\mathrm{H}_\mathrm{I}}=\frac{1}{1+\frac{\Gamma}{\alpha_{\mathrm{rec,H}_\mathrm{II}}n_\mathrm{e}}}}$ (6)

## $\bullet$ $\text{Caso 2}$

## $\text{Supongamo que para un tiempo mayor que cero no se tiene equilibrio de fotoionización.}$ 

## $\text{Derivando (1) y sustituyendo (2) y (3) se tiene:}$ 

## $\boxed{\ddot{n}_{\mathrm{H}_\mathrm{I}}(t)=-\dot\Gamma_{\mathrm{H}_\mathrm{I}}(t)n_{\mathrm{H}_\mathrm{I}}(t)-\Gamma_{\mathrm{H}_\mathrm{I}}(t)\dot n_{\mathrm{H}_\mathrm{I}}(t)
+\alpha_{\mathrm{rec,H}_\mathrm{II}}n_\mathrm{e}\Gamma_{\mathrm{H}_\mathrm{I}}(t)n_{\mathrm{H}_\mathrm{I}}(t)-\alpha_{\mathrm{rec,H}_\mathrm{II}}^2n_\mathrm{e}^2+\alpha_{\mathrm{rec,H}_\mathrm{II}}^2n_\mathrm{e}^2n_{\mathrm{H}_\mathrm{I}}(t)}$ (7)

## $\text{Volviendo a tomar } \Gamma \text{= cte, (7) se reduce a:}$

## $\boxed{\ddot{n}_{\mathrm{H}_\mathrm{I}}(t)=-\Gamma_{\mathrm{H}_\mathrm{I}}\dot n_{\mathrm{H}_\mathrm{I}}(t)
+\alpha_{\mathrm{rec,H}_\mathrm{II}}n_\mathrm{e}\Gamma_{\mathrm{H}_\mathrm{I}}n_{\mathrm{H}_\mathrm{I}}(t)-\alpha_{\mathrm{rec,H}_\mathrm{II}}^2n_\mathrm{e}^2+\alpha_{\mathrm{rec,H}_\mathrm{II}}^2n_\mathrm{e}^2n_{\mathrm{H}_\mathrm{I}}(t)}$ (8)

## $\text{Sea: } \boxed{\aleph=\alpha_{\mathrm{rec,H}_\mathrm{II}}n_\mathrm{e}}$ (9)

## $\text{Sustituyendo (9) en (8) queda:}$

## $\boxed{\ddot{n}_{\mathrm{H}_\mathrm{I}}(t)=-\Gamma_{\mathrm{H}_\mathrm{I}}\dot n_{\mathrm{H}_\mathrm{I}}(t)
+\aleph\left(\aleph+\Gamma_{\mathrm{H}_\mathrm{I}}\right)n_{\mathrm{H}_\mathrm{I}}(t)-\aleph^2}$ (10)

## $\text{Voy a reescribir (10) redefiniendo variables para que quede más claro}$

## $\boxed{\ddot{x}(t)=-\Gamma \dot x (t) +\aleph\left(\aleph+\Gamma\right)x(t)-\aleph^2}$ (11)

## $\text{donde}$

## $\Gamma = \Gamma_{\mathrm{H}_\mathrm{I}}$

## $x(t)=n_{\mathrm{H}_\mathrm{I}}(t)$

## $x(t): \text{solución general}$

## $x_p(t): \text{solución particular}$

## $x_c(t): \text{solución complementaria}$

## $\text{Ahora vamos a resolver (11), donde sabemos que su solución es:}$

## $x(t)= x_p(t) +x_c(t)$

## $\text{Entonces, para encontrar la solución complementaria vamos a resolver:}$

## $\boxed{\ddot{x}_c(t)+\Gamma \dot x_c (t) -\aleph\left(\aleph+\Gamma\right)x_c(t)=0}$ (12)

## $\text{Proponemos como solución a:}$

## $x_c(t)=e^{\lambda t}$

## $\text{Sustituyendo y resolviendo nos queda:}$

## $\boxed{\lambda^2+\Gamma \lambda -\aleph\left(\aleph+\Gamma\right)=0}$ (13)

## $\text{Con:}$

## $\boxed{\lambda=\frac{-\Gamma\pm\sqrt{\Gamma^2+4\aleph\left(\aleph+\Gamma\right)}}{2}}$ 

## $\text{Que se reduce a:}$

## $\boxed{\lambda=\frac{-\Gamma\pm\sqrt{\left(\Gamma+2\aleph\right)^2}}{2}}$ 

## $\text{Que se reduce a:}$

## $\boxed{\lambda=\frac{-\Gamma\pm\left(\Gamma+2\aleph\right)}{2}}$ 

## $\text{Tal que:}$

## $\boxed{\lambda_1=\aleph}$ (14)

## $\boxed{\lambda_2=-\left(\Gamma+\aleph\right)}$ (15)

## $\text{Como queremos que para tiempos infinitos } x(t) \text{ no diverga}$

## $\text{Tenemos que la solución complementaria es:}$

## $\boxed{x_c(t)=e^{-t\left(\Gamma+\aleph\right)}}$ (16)

## $\text{Ahora resolvemos el caso particular: } x_p(t)$

## $\text{Supongamos que } x_p(t)=k \text{ donde } k  \text{ es una constante, tal que}$

## $\boxed{\ddot{x}_p(t)+\Gamma \dot x_p (t) -\aleph\left(\aleph+\Gamma\right)x_p(t)+\aleph^2=0}$ (17)

## $\text{se reduce a:}$

## $\boxed{-\aleph\left(\aleph+\Gamma\right)k+\aleph^2=0}$ (18)

## $\text{Tal que:}$

## $\boxed{k=\frac{\aleph}{\aleph+\Gamma}}$ (19)

## $\text{Así, la solución particular es:}$

## $\boxed{x_p(t)=\frac{\aleph}{\aleph+\Gamma}}$ (20)

## $\text{Por lo tanto la solución general es:}$

## $\boxed{x(t)=e^{-t\left(\Gamma+\aleph\right)}+\frac{\aleph}{\aleph+\Gamma}}$ (21)

## $\therefore$

## $\boxed{{n}_{\mathrm{H}_\mathrm{I}}(t)= \left \{ \begin{matrix} \frac{\aleph}{\aleph+\Gamma} & \mbox{ si } & t\leq0
\\ e^{-t\left(\Gamma+\aleph\right)}+\frac{\aleph}{\aleph+\Gamma} & \mbox{ si } & t > 0\end{matrix}\right.}$ (22)

## $\text{Donde reescribí (6) en términos de (9), es decir:}$ 

## $\boxed{\aleph=\alpha_{\mathrm{rec,H}_\mathrm{II}}n_\mathrm{e}}$ 

## $\text{Ahora Implementamos (3) y (22)}$ 


```julia
function nHI(tiempo,ne)
    
α_rec_HII=1
    
ℵ=α_rec_HII+ne
Γ=1
    
ts=Float64[]
nHIs=Float64[]
nHIIs=Float64[]
Fs=Float64[]
    
for t in -tiempo:tiempo/10000:tiempo
        
       if t<=0

            densidad_HI=ℵ/(ℵ+Γ)
            densidad_HII=1-densidad_HI
            push!(nHIIs,densidad_HII)
            push!(nHIs,densidad_HI)
            push!(ts,t)  
            push!(Fs,Γ)
            else

            densidad_HI=exp(-t*(ℵ+Γ))+(ℵ/(ℵ+Γ))
            densidad_HII=1-densidad_HI
            push!(nHIIs,densidad_HII)
            push!(nHIs,densidad_HI)
            push!(ts,t)   
            push!(Fs,Γ)
        end
        

    end
    ts,nHIs,nHIIs,Fs
end
```
