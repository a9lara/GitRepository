## $\boxed{\dot{n}_{\mathrm{H}_\mathrm{I}}(t)=-\Gamma_{\mathrm{H}_\mathrm{I}}(t)n_{\mathrm{H}_\mathrm{I}}(t)+\alpha_{\mathrm{rec,H}_\mathrm{II}}n_\mathrm{e}n_{\mathrm{H}_\mathrm{II}}(t)}$ (1)

## $\boxed{\dot{n}_{\mathrm{H}_\mathrm{II}}(t)=\Gamma_{\mathrm{H}_\mathrm{I}}(t)n_{\mathrm{H}_\mathrm{I}}(t)-\alpha_{\mathrm{rec,H}_\mathrm{II}}n_\mathrm{e}n_{\mathrm{H}_\mathrm{II}}(t)}$ (2)

## $\text{Suponemos:}$

## $n_\mathrm{e}\approx\text{cte}$

## $N_\mathrm{H}=N_{\mathrm{H}_\mathrm{I}}(t)+N_{\mathrm{H}_\mathrm{II}}(t)$

## $\text{Normalizamos haciendo}$ 

## $n_\mathrm{H}=\frac{N_\mathrm{H}}{N_\mathrm{H}}$

## $n_{\mathrm{H}_\mathrm{I}}(t)=\frac{N_{\mathrm{H}_\mathrm{I}}(t)}{N_\mathrm{H}}$

## $n_{\mathrm{H}_\mathrm{II}}(t)=\frac{N_{\mathrm{H}_\mathrm{II}}(t)}{N_\mathrm{H}}$



## $\text{Tal que}$

## $n_{\mathrm{H}_\mathrm{I}}(t)+n_{\mathrm{H}_\mathrm{II}}(t)=1$ 

## $\boxed{n_{\mathrm{H}_\mathrm{II}}(t)=1-n_{\mathrm{H}_\mathrm{I}}(t)}$ (3)

## $\text{Derivando (1) y sustituyendo (2) y (3) se tiene:}$ 

## $\boxed{\ddot{n}_{\mathrm{H}_\mathrm{I}}(t)=-\dot\Gamma_{\mathrm{H}_\mathrm{I}}(t)n_{\mathrm{H}_\mathrm{I}}(t)-\Gamma_{\mathrm{H}_\mathrm{I}}(t)\dot n_{\mathrm{H}_\mathrm{I}}(t)
+\alpha_{\mathrm{rec,H}_\mathrm{II}}n_\mathrm{e}\Gamma_{\mathrm{H}_\mathrm{I}}(t)n_{\mathrm{H}_\mathrm{I}}(t)-\alpha_{\mathrm{rec,H}_\mathrm{II}}^2n_\mathrm{e}^2+\alpha_{\mathrm{rec,H}_\mathrm{II}}^2n_\mathrm{e}^2n_{\mathrm{H}_\mathrm{I}}(t)}$ (4)

## $\text{Tomando } \Gamma \text{= cte, (4) se reduce a:}$

## $\boxed{\ddot{n}_{\mathrm{H}_\mathrm{I}}(t)=-\Gamma_{\mathrm{H}_\mathrm{I}}\dot n_{\mathrm{H}_\mathrm{I}}(t)
+\alpha_{\mathrm{rec,H}_\mathrm{II}}n_\mathrm{e}\Gamma_{\mathrm{H}_\mathrm{I}}n_{\mathrm{H}_\mathrm{I}}(t)-\alpha_{\mathrm{rec,H}_\mathrm{II}}^2n_\mathrm{e}^2+\alpha_{\mathrm{rec,H}_\mathrm{II}}^2n_\mathrm{e}^2n_{\mathrm{H}_\mathrm{I}}(t)}$ (5)

## $\text{Sea: } \boxed{\aleph=\alpha_{\mathrm{rec,H}_\mathrm{II}}n_\mathrm{e}}$ (6)

## $\text{Sustituyendo (6) en (5) queda:}$

## $\boxed{\ddot{n}_{\mathrm{H}_\mathrm{I}}(t)=-\Gamma_{\mathrm{H}_\mathrm{I}}\dot n_{\mathrm{H}_\mathrm{I}}(t)
+\aleph\left(\aleph+\Gamma_{\mathrm{H}_\mathrm{I}}\right)n_{\mathrm{H}_\mathrm{I}}(t)-\aleph^2}$ (7)

## $\text{Voy a reescribir (7) redefiniendo variables para que quede más claro}$

## $\boxed{\ddot{x}(t)=-\Gamma \dot x (t) +\aleph\left(\aleph+\Gamma\right)x(t)-\aleph^2}$ (8)

## $\text{donde}$

## $\Gamma = \Gamma_{\mathrm{H}_\mathrm{I}}$

## $x(t)=n_{\mathrm{H}_\mathrm{I}}(t)$

## $x(t): \text{solución general}$

## $x_p(t): \text{solución particular}$

## $x_c(t): \text{solución complementaria}$

## $\text{Ahora vamos a resolver (8), donde sabemos que su solución es:}$

## $x(t)= x_p(t) +x_c(t)$

## $\text{Entonces, para encontrar la solución complementaria vamos a resolver:}$

## $\boxed{\ddot{x}_c(t)+\Gamma \dot x_c (t) -\aleph\left(\aleph+\Gamma\right)x_c(t)=0}$ (9)

## $\text{Proponemos como solución a:}$

## $x_c(t)=Ae^{\lambda t}$

## $\text{Con } A \text{ una cconstante. Sustituyendo y resolviendo nos queda:}$

## $\boxed{\lambda^2+\Gamma \lambda -\aleph\left(\aleph+\Gamma\right)=0}$ (10)

## $\text{Con:}$

## $\boxed{\lambda=\frac{-\Gamma\pm\sqrt{\Gamma^2+4\aleph\left(\aleph+\Gamma\right)}}{2}}$ 

## $\text{Que se reduce a:}$

## $\boxed{\lambda=\frac{-\Gamma\pm\sqrt{\left(\Gamma+2\aleph\right)^2}}{2}}$ 

## $\text{Que se reduce a:}$

## $\boxed{\lambda=\frac{-\Gamma\pm\left(\Gamma+2\aleph\right)}{2}}$ 

## $\text{Tal que:}$

## $\boxed{\lambda_1=\aleph}$ (11)

## $\boxed{\lambda_2=-\left(\Gamma+\aleph\right)}$ (12)

## $\text{Como queremos que para tiempos infinitos } x(t) \text{ no diverga,}$

## $\text{tenemos que la solución complementaria es:}$

## $\boxed{x_c(t)=Ae^{-t\left(\Gamma+\aleph\right)}}$ (13)

## $\text{Con } A \text{ una constante.}$

## $\text{Ahora resolvemos el caso particular: } x_p(t)$

## $\text{Supongamos que } x_p(t)=k \text{ donde } k  \text{ es una constante, tal que}$

## $\boxed{\ddot{x}_p(t)+\Gamma \dot x_p (t) -\aleph\left(\aleph+\Gamma\right)x_p(t)+\aleph^2=0}$ (14)

## $\text{se reduce a:}$

## $\boxed{-\aleph\left(\aleph+\Gamma\right)k+\aleph^2=0}$ (15)

## $\text{Tal que:}$

## $\boxed{k=\frac{\aleph}{\aleph+\Gamma}}$ (16)

## $\text{Así, la solución particular es:}$

## $\boxed{x_p(t)=\frac{\aleph}{\aleph+\Gamma}}$ (17)

## $\text{Por lo tanto la solución general es:}$

## $\boxed{x(t)=Ae^{-t\left(\Gamma+\aleph\right)}+\frac{\aleph}{\aleph+\Gamma}}$ (18)

## $\text{Sabiendo la condición la condición de frontera tenemos:}$

## $\boxed{n_{\mathrm{H}_\mathrm{I}}(t=t_1)=n_{\mathrm{H}_{\mathrm{I}_\mathrm{o}}}}$ (19)

## $\text{Tal que:}$

## $\boxed{n_{\mathrm{H}_{\mathrm{I}_\mathrm{o}}}=Ae^{-t\left(\Gamma+\aleph\right)}+\frac{\aleph}{\aleph+\Gamma}}$ (20)

## $\text{Despejando } A \text{ se obtiene:}$

## $\boxed{A=\left(n_{\mathrm{H}_{\mathrm{I}_\mathrm{o}}}-\frac{\aleph}{\aleph+\Gamma}\right)e^{t_1\left(\Gamma+\aleph\right)}}$ (21)

## $\text{Así, sustituyendo en (21) obtenemos:}$

## $\boxed{n_{\mathrm{H}_\mathrm{I}}(t)=\left[\left(n_{\mathrm{H}_{\mathrm{I}_\mathrm{o}}}-\frac{\aleph}{\aleph+\Gamma}\right)e^{t_1\left(\Gamma+\aleph\right)}\right]e^{-t\left(\Gamma+\aleph\right)}+\frac{\aleph}{\aleph+\Gamma}}$ (22)

## $\text{Que se puede ver como:}$

## $\boxed{n_{\mathrm{H}_\mathrm{I}}(t)=\left(n_{\mathrm{H}_{\mathrm{I}_\mathrm{o}}}-\frac{\aleph}{\aleph+\Gamma}\right)e^{\left(t_1-t\right)\left(\Gamma+\aleph\right)}+\frac{\aleph}{\aleph+\Gamma}}$ (23)

## $\text{Ahora vamos a buscar la condición densidad de hidrógeno neutro cuando el gas está en equilibrio.}$

## $\text{La tasa de fotoionización, }$ $\Gamma , $  $\text{es la probabilidad de tener una ionización a partir de una frecuencia }$ $\nu$ $\text{con valores que van desde la frecuencia de ionización, }$ $\nu_o ,$ $\text{ hasta}$ $\infty$. $\text{La tasa de fotoionización se expresa como:}$

## $\boxed{\Gamma=\displaystyle{\int_{\nu_o}^{\infty}}\frac{L_{\nu}}{4\pi r^2 h \nu}a_{\nu}d\nu}$ (24)

## $\text{Si tenemos que:}$

## $L_{\nu}=L_{\nu_o} \left(\frac{\nu}{\nu_o}\right)^{n}$

## $a_{\nu} =a_{\nu_o} \left(\frac{\nu}{\nu_o}\right)^{m}$

## $\text{Nos queda que (1) se puede ver como:}$

## $\boxed{\Gamma = \displaystyle{ \frac{L_{\nu_o}a_o}{4 \pi r^2h}  \int_{\nu_o}^{\infty}  \left(\frac{\nu}{\nu_o}\right)^{n}                  {\nu}^{-1}       \left(\frac{\nu}{\nu_o}\right)^{m}          d\nu}}$ (25)

## $\text{Que se reduce a:}$

## $\boxed{\Gamma = \frac{L_{\nu_o}a_o\nu_o^{-(n+m)}}{4 \pi r^2h}\displaystyle{\left(\int_{\nu_o}^{\infty}  \nu^{n+m-1} d\nu\right)}}$ (26)

## $\text{Integrando queda:}$

## $\boxed{\Gamma =\displaystyle{\left(\frac{1}{-(n+m)}\right)\left(\frac{L_{\nu_o}a_o}{4 \pi r^2h}\right)}}$ (27)

## $\text{Para el caso de gas en equilibrio de fotoionización, considerando (1) = 0 y (27) tenemos:}$

## $\boxed{\displaystyle{\left(\frac{1}{-(n+m)}\right)\left(\frac{L_{\nu_o}a_o}{4 \pi r^2h}\right)=\frac{N_pN_e\alpha_\mathrm{rec}}{N_{\mathrm{H}_\mathrm{I}}}}}$ (28)

## $\text{Para el caso del hidrógeno se tiene que: } N_p=N_e=N_{\mathrm{H II}_\mathrm{o}}$

## $\text{Por lo que reescribiendo (28) se tiene:}$

## $\boxed{\displaystyle{\left(\frac{N_{\mathrm{H I}_\mathrm{o}}}{-(n+m)}\right)\left(\frac{L_{\nu_o}a_o}{4 \pi r^2h}\right)=N_{\mathrm{H II}_\mathrm{o}}^2\alpha_\mathrm{rec}}}$ (29)

## $\text{Normalizando tenemos: }$

## $\boxed{N_{\mathrm{H I}_\mathrm{o}}=N_{\mathrm{H}}n_{\mathrm{H I}_\mathrm{o} }}$ (30)

## $\text{Tal que: }$

## $\boxed{N_{\mathrm{H II}_\mathrm{o}}= N_{\mathrm{H}}(1-n_{\mathrm{H I}_\mathrm{o}}) }$ (31)

## $\text{Así, reescribiendo (29) tenemos: }$

## $\boxed{\displaystyle{\left(\frac{N_{\mathrm{H}}n_{\mathrm{H I}_\mathrm{o}}}{-(n+m)}\right)\left(\frac{L_{\nu_o}a_o}{4 \pi r^2h}\right)=N_{\mathrm{H}}^2(1-n_{\mathrm{H I}_\mathrm{o}})^2\alpha_\mathrm{rec}}}$ (32)

## $\text{Reescribiendo (32) tenemos: }$

## $\boxed{\displaystyle{\left(\frac{n_{\mathrm{H I}_\mathrm{o}}}{-(n+m)}\right)\left(\frac{L_{\nu_o}a_o}{4 \pi r^2h}\right)=N_{\mathrm{H}}(1-n_{\mathrm{H I}_\mathrm{o}})^2\alpha_\mathrm{rec}}}$ (33)

## $\text{Reescribiendo (33) tenemos: }$

## $\boxed{\displaystyle{n_{\mathrm{H I}_\mathrm{o}}\left[\left(\frac{1}{-(n+m)N_{\mathrm{H}}\alpha_\mathrm{rec}}\right)\left(\frac{L_{\nu_o}a_o}{4 \pi r^2h}\right)\right]=(1-n_{\mathrm{H I}_\mathrm{o}})^2}}$ (34)

## $\text{Reescribiendo (34) tenemos: }$

## $\boxed{\displaystyle{n_{\mathrm{H}_\mathrm{I}}\left[\left(\frac{1}{-(n+m)N_{\mathrm{H}}\alpha_\mathrm{rec}}\right)\left(\frac{L_{\nu_o}a_o}{4 \pi r^2h}\right)\right]=1-2n_{\mathrm{H I}_\mathrm{o}}+n_{\mathrm{H I}_\mathrm{o}}^2}}$ (35)

## $\text{Que nos da la siguiente ecuación que hay que resolver para } n_{\mathrm{H}_\mathrm{I}}$

## $\boxed{\displaystyle{n_{\mathrm{H I}_\mathrm{o}}^2-n_{\mathrm{H I}_\mathrm{o}}\left(2+\left[\left(\frac{1}{-(n+m)N_{\mathrm{H}}\alpha_\mathrm{rec}}\right)\left(\frac{L_{\nu_o}a_o}{4 \pi r^2h}\right)\right]\right)+1=0}}$ (36)

## $\text{Resolviendo (36) se obtiene que:}$

## $n_{\mathrm{H I}_\mathrm{o}}=\frac{(B+2)-\sqrt{B^2+4B}}{2}$ (37)

## $\text{donde:}$

## $B=\left[\left(\frac{1}{-(n+m)N_{\mathrm{H}}\alpha_\mathrm{rec}}\right)\left(\frac{L_{\nu_o}a_o}{4 \pi r^2h}\right)\right]$ (38)

## $\text{Tal que la fracción de hidrógeno neutro cuando el gas está en equilibrio es:}$

## $n_{\mathrm{H I}_\mathrm{o}}=\frac{\left(\left[\left(\frac{1}{-(n+m)N_{\mathrm{H}}\alpha_\mathrm{rec}}\right)\left(\frac{L_{\nu_o}a_o}{4 \pi r^2h}\right)\right]+2\right)-\sqrt{\left[\left(\frac{1}{-(n+m)N_{\mathrm{H}}\alpha_\mathrm{rec}}\right)\left(\frac{L_{\nu_o}a_o}{4 \pi r^2h}\right)\right]^2+4\left[\left(\frac{1}{-(n+m)N_{\mathrm{H}}\alpha_\mathrm{rec}}\right)\left(\frac{L_{\nu_o}a_o}{4 \pi r^2h}\right)\right]}}{2}$ (39)

## $\text{Tomando:}$

## $n=-2.5$

## $m=-3$

## $L_{\nu_o}=4.4\times 10 ^{45} \mathrm{erg}$

## $a_o=5\times 10 ^{-18} \mathrm{cm}^2$

## $r=3.086\times 10 ^{18} \mathrm{cm}$

## $h=6.6626\times 10 ^{-27} \mathrm{erg} \ \mathrm{s}$

## $\alpha_{\mathrm{rec}}=4\times 10 ^{-13} \mathrm{cm}^{-3} \mathrm{s}^{-1} $

## $\text{Tal que:}$

## $\Gamma_o=5\times 10 ^{15}  \mathrm{cm}^{-3} \mathrm{s}^{-1}$


```julia
# Implementamos

function ξ(tf,nH,a)
    
    # tf es el tiempo durante el cual vamos a dejar que se fotoionice el gas
    # nH es el número de átomos de hidrógeno total en el gas
    # a es el factor por el cuál cambia la tasa de fotoionización al tiempo t1
    
    # Valores de los parámetros
    
    Lν0 = 4.4e45
    a0 = 5e-18
    r = 3.086e18
    h = 6.6626e-27
    n = -2.5
    m = -3
    α_rec = 4e-13
    t1 = tf/2
    ti = 0
    Δt = tf/1000
           
    #  Calculamos la densidad de HI inicial (nHI0) cuando el gas está en equilibrio implementando (39)
    
        # Sea ℶ igual a la constante:
    
    ℶ = (Lν0*a0)/(4*π*h*r^2)*((1)/(-(n+m)*α_rec))
    
        # Tal que (39) es:
    
    nHI0 = (1/2)*(((ℶ/nH)+2)-sqrt(((ℶ/nH))^2+4*(ℶ/nH)))
    
    #  Calculamos la fracción de hidrógeno ionizado inicial (nHII) dada nHIO
    
    nHII = 1 - nHI0
    
    #  La tasa de fotoionización inicial es: 
    
    Γ0 = (Lν0*a0)/(4*π*h*r^2)*((1)/(-(n+m)))
    
    #  Determinamos L_max para normalizar L
    
    Γmax = max((Γ0,a*Γ0)...)
    
    # Definimos los arrays donde guardaremos nuestros datos
    
    nHIs = Float64[]
    nHIIs = Float64[]
    Γs = Float64[]
    ts = Float64[]
            
    # Para este caso, el número de electrones líbres (ne) va a ser constante
        
    ne_constante = nHII*nH
        
    ℵ = ne_constante*α_rec
        
    for t in ti:Δt:tf
        
        if t < t1
            
            Γ = Γ0
            nHI = nHI0
            nHII = 1-nHI 
            
        else
            
            Γ = a*Γ0
            nHI = ((nHI0-(ℵ/(ℵ+Γ)))*(exp((t1-t)*(Γ+ℵ))))+(ℵ/(ℵ+Γ))    
            nHII = 1-nHI 
            
        end
        
        push!(ts,t)
        push!(nHIs,nHI)
        push!(nHIIs,nHII)
        push!(Γs,Γ/Γmax)
            
    end
    
    ts,nHIs,nHIIs,Γs
    
end
```

```julia
# Evaluamos

tiempo = 1e10 # segundos
número_de_átomos_de_hidrógeno = [100,1000,10000] # partículas / cm^{3}
factor_que_multiplica_Γ=1e-25

t_1,nHI_1,nHII_1,L_1 = ξ(tiempo,número_de_átomos_de_hidrógeno[1],factor_que_multiplica_Γ)
t_2,nHI_2,nHII_2,L_2 = ξ(tiempo,número_de_átomos_de_hidrógeno[2],factor_que_multiplica_Γ)
t_3,nHI_3,nHII_3,L_3 = ξ(tiempo,número_de_átomos_de_hidrógeno[3],factor_que_multiplica_Γ);
```

```julia
# Graficamos

using PyPlot

nH_1=número_de_átomos_de_hidrógeno[1]
nH_2=número_de_átomos_de_hidrógeno[2]
nH_3=número_de_átomos_de_hidrógeno[3]

title("Al la mitad del tiempo, Γ aumenta a-veces",fontsize=18)

plot(t_1,nHI_1,"-b",label="nH=$nH_1")
plot(t_1,nHII_1,":b")

plot(t_2,nHI_2,"-g",label="nH=$nH_2")
plot(t_2,nHII_2,":g")

plot(t_3,nHI_3,"-y",label="nH=$nH_3")
plot(t_3,nHII_3,":y")

plot(t_3,L_3,"-k",label="a=$factor_que_multiplica_Γ")

xticks(fontsize=18)
yticks(fontsize=18)
xlabel(L"t \ [\mathrm{s}]",fontsize=18)
ylabel(L"n_{HI} \ [\mathrm{cm}^{-3}]",fontsize=18)
legend(fontsize=18,bbox_to_anchor=(1, .56),loc="lower left");
```
